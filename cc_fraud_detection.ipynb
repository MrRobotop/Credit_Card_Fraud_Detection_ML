```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
cc_data = pd.read_csv('dataset/creditcard.csv')
```

```python
cc_data['Class'].value_counts()
```

```python
#This looks like an imbalanced dataset
normal = cc_data[cc_data.Class == 0]
fraud = cc_data[cc_data.Class == 1]
```

```python
normal.shape
fraud.shape
```

```python
normal.shape
```

```python
cc_data.groupby('Class').mean()
```

```python
#We are going to undersample by building a similar distribution of normal transaction of 492 values aand concatenate it

normal_sample = normal.sample(n=492)

cc_data_new = pd.concat([normal_sample, fraud], axis=0)
```

```python
cc_data_new.shape
```

```python
cc_data_new.head()
```

```python
cc_data_new.groupby('Class').mean()
```

```python
x = cc_data_new.drop(columns='Class', axis=1)
y = cc_data_new['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify = y, random_state=8)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
train_pred = model.predict(x_train)
test_pred = model.predict(x_test)

train_accuracy = accuracy_score(train_pred, y_train)
test_accuracy = accuracy_score(test_pred, y_test)

print('Training Accuracy: ', train_accuracy)
print('Testing Accuracy', test_accuracy)
```
